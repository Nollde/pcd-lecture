```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from cycler import cycler
```

```python
def plot2d(ul, ur, ll, lr, weights_lr=None):
    d0 = 0
    for d1 in range(1, N_DIMS):
        fig, axs = plt.subplots(2, 2, figsize=(8, 6))
        opts = dict(bins=20, range=[[0, SCALE], [0, SCALE]], density=True)
        axs[0,0].hist2d(ul[:, d0], ul[:, d1], **opts)
        axs[0,1].hist2d(ur[:, d0], ur[:, d1], **opts)
        axs[1,0].hist2d(ll[:, d0], ll[:, d1], **opts)
        h = axs[1,1].hist2d(lr[:, d0], lr[:, d1], weights=weights_lr, **opts)
        fig.colorbar(h[3], ax=axs.flat)
        fig.show()    
```

```python
N_PEAK = 15
N_DIMS = 2
SCALE = 20
```

```python
rng = np.random.default_rng(123)
centers_ac = rng.uniform(0, SCALE, size=(N_PEAK, N_DIMS))
centers_bd = rng.uniform(0, SCALE, size=(N_PEAK, N_DIMS))
```

```python
def make_samples(dest: bool, apply: bool) -> np.ndarray:
    # dest: is this the destination region? (B/D) expected to influnece distribution
    # apply: is this the application region? (C/D) should not influnece distribution
    centers = centers_bd if dest else centers_ac
    
    samples = 1_000_000
    # simulate poor statistics in target region (D)
    if dest and apply:
        samples = 1_000
    
    output = []
    for i, center in enumerate(centers):
        output.append(rng.normal(
            loc=center,
            scale=(SCALE + i) / (1 + i),
            size=(samples // N_PEAK, N_DIMS)
        ))
    return np.concatenate(output)
```

```python
a = make_samples(False, False)
b = make_samples(True, False)
c = make_samples(False, True)
d = make_samples(True, True)
```

```python
plot2d(a, b, c, d)
```

```python
model = tf.keras.Sequential(
    layers=[
        tf.keras.Input(shape=(N_DIMS,)),
        tf.keras.layers.Dense(256, activation="relu"),
        tf.keras.layers.Dense(256, activation="relu"),
        tf.keras.layers.Dense(256, activation="relu"),
        tf.keras.layers.Dense(1),
    ]
)
```

```python
bce = tf.keras.losses.BinaryCrossentropy(from_logits=True)
model.compile(optimizer="adam", loss=bce, metrics=["accuracy"])
# model.summary()
```

```python
x = np.concatenate([a, b])
y = np.concatenate([np.ones(len(a)), np.zeros(len(b))])
```

```python
p = np.random.permutation(len(x))
x, y = x[p], y[p]
```

```python
model.fit(x=x, y=y, batch_size=1024, epochs=3, validation_split=0.2, verbose=2)
```

```python
weights_c_to_d = np.exp(-model.predict(c, batch_size=4096)[:, 0])
```

```python
plot2d(a, b, c, lr=c, weights_lr=weights_c_to_d)
```

```python
# 1-D projections
for d0 in range(N_DIMS):
    fig, axs = plt.subplots(figsize=(10,4))
    for opts in cycler(
        x=[a, b, c, d, c],
        weights=[None] * 4 + [weights_c_to_d],
        label=["A","B","C","D","reweighted"],
        linestyle=["--", "--", ":", ":", "-"],
    ):
        axs.hist(
            opts.pop("x")[:, d0],
            **opts,
            bins=20,
            range=[0, SCALE],
            density=True,
            histtype="step",
            lw=3,
        )
    axs.legend(loc=0)
    fig.show()
```
